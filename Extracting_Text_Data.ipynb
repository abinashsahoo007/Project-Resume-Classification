# **Extracting text from DataSet**

```python
# extracting data from peoplesoft resumes
file1=[]
category1=[]
directory1="/content/drive/MyDrive/Resumes_Docs/Peoplesoft resumes"
for i in os.listdir(directory1):
  if i.endswith('.docx'):
    os.path.join(directory1,i)
    text=(textract.process(os.path.join(directory1,i)).decode('utf-8'))
    file1.append(text)
    category1.append('PeopleSoft')
```

```python
data1=pd.DataFrame(file1,columns=['raw_text'])
data1['category']=category1
data1
```

```python
# extracting data from react developer resumes
file2=[]
category2=[]
directory2="/content/drive/MyDrive/Resumes_Docs/React Developer resumes"
for i in os.listdir(directory2):
  if i.endswith(".docx"):
    os.path.join(directory2,i)
    text=(docx2txt.process(os.path.join(directory2,i)))
    file2.append(text)
    category2.append('React Developer')
```

```python
data2=pd.DataFrame(file2,columns=['raw_text'])
data2['category']=category2
data2
```

```python
# extracting data from sql developer resumes
file3=[]
category3=[]
directory3="/content/drive/MyDrive/Resumes_Docs/SQL Developer resumes"
for i in os.listdir(directory3):
  if i.endswith(".docx"):
    os.path.join(directory3,i)
    text=(textract.process(os.path.join(directory3,i)).decode('utf-8'))
    file3.append(text)
    category3.append('SQL Developer')
```

```python
data3=pd.DataFrame(file3,columns=['raw_text'])
data3['category']=category3
data3
```

```python
# extracting data from workday resumes
file4=[]
category4=[]
directory4="/content/drive/MyDrive/Resumes_Docs/workday resumes"
for i in os.listdir(directory4):
  if i.endswith('.docx'):
    os.path.join(directory4,i)
    text=(docx2txt.process(os.path.join(directory4,i)))
    file4.append(text)
    category4.append('Workday')
```

```python
data4=pd.DataFrame(file4,columns=['raw_text'])
data4['category']=category4
data4
```

# **Final DataSet**

**Combining all data frame to form one complete new data frame**

```python
# Now combining all the raw datarame to form one dataframe
```

```python
#concatate
resume_df=pd.concat([data1,data2,data3,data4],axis=0)
resume_df.reset_index(drop=True,inplace=True)
```

```python
#renaming the columns names
resume_df.rename(columns={'raw_text':'Resumes','category':'Category'},inplace=True)
resume_df=resume_df[['Category','Resumes']]
resume_df
```
